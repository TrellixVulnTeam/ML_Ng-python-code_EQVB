```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

下面是画sigmoid函数的





```python
X = np.linspace(-10,10,100)
Y = 1/(1+np.exp(-1*X))
fig, ax = plt.subplots(figsize = (12,9))
ax.plot(X, Y, 'r')
plt.show()
```

下面是画cost函数的




```python
X = np.linspace(0,1,100)
Y = -1 * np.log10(X)
fig, ax = plt.subplots(figsize = (12,9))
ax.plot(X, Y, 'r')
plt.title('y=1')
plt.show()

X = np.linspace(0,1,100)
Y = -1 * np.log(1 - X)
fig, ax = plt.subplots(figsize = (12,9))
ax.plot(X, Y, 'r')
plt.title('y=0')
plt.show()
```
